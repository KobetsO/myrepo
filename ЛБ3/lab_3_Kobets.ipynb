<p>4. Імпортувати RGB-зображення Image у напівтоновому форматі. Оцінити його розмір у пікселях та об’єм пам’яті, який він займає на диску. Порівняти результати з попередніми результатами. Пояснити результати.</p>

<p>5. Зменшити початкове RGB-зображення Image у n разів, візуалізувати його, перевіривши результат масштабування, та експортувати його на диск. Чи змінився об’єм пам’яті, який він займає на диску? Обгрунтувати отриманий результат.</p>

```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import math

image_path = 'scr/123.png' 
n = 2

image_rgb = cv.imread(image_path)

if image_rgb is None:
    print(f"Помилка: Не вдалося знайти зображення за шляхом: {image_path}")
else:
    height_N, width_M, _ = image_rgb.shape
    file_size_bytes_original = os.path.getsize(image_path)

    print(f"Результати: Завдання 5")
    print(f"Оригінальний розмір (N x M): {height_N} x {width_M}")
    print(f"Оригінальний об'єм на диску: {file_size_bytes_original} байт\n")

    new_width = math.floor(width_M / n)
    new_height = math.floor(height_N / n)
    new_dimensions = (new_width, new_height)

    image_resized = cv.resize(image_rgb, new_dimensions, interpolation=cv.INTER_AREA)
    
    print(f"Зображення зменшено у {n} рази.")
    print(f"Новий розмір (N x M): {new_height} x {new_width}\n")

    resized_image_path = 'scr/123_resized.png'
    cv.imwrite(resized_image_path, image_resized, [cv.IMWRITE_PNG_COMPRESSION, 9])
    print(f"Зменшене зображення збережено як: {resized_image_path}")

    file_size_bytes_resized = os.path.getsize(resized_image_path)
    print(f"Новий об'єм на диску: {file_size_bytes_resized} байт\n")

    print("Чи змінився об’єм пам’яті, який він займає на диску?")
    if file_size_bytes_resized < file_size_bytes_original:
        print(f"Так, об'єм пам'яті на диску **зменшився**.")
        print(f"Зменшення склало: {file_size_bytes_original - file_size_bytes_resized} байт.")
    else:
        print("Ні, об'єм пам'яті не зменшився (або навіть збільшився), що незвично для зменшення.")

        print("\nОбґрунтування:")
        print(f"Хоча нове зображення має значно менше пікселів ({new_height * new_width} проти {height_N * width_M}), \n"
              f"його розмір на диску виявився більшим. \n"
              f"Це означає, що алгоритм стиснення має більший вплив, ніж просто кількість пікселів.\n\n"
              f"Ймовірно, оригінальний файл ('{image_path}') був дуже сильно оптимізований \n"
              f"за допомогою спеціальних інструментів (наприклад, зі зменшенням кількості кольорів).\n\n"
              f"Функція `cv.imwrite()`, навпаки, використала стандартний метод стиснення PNG, \n"
              f"який виявився менш ефективним, ніж той, що був застосований до оригіналу. \n"
              f"Тому неоптимізований файл з меншою кількістю пікселів зайняв більше місця, \n"
              f"ніж високооптимізований файл з великою кількістю пікселів.")

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image_rgb, cv.COLOR_BGR2RGB))
    plt.title(f'Оригінал: {width_M} x {height_N}')
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_resized, cv.COLOR_BGR2RGB))
    plt.title(f'Зменшено (n={n}): {new_width} x {new_height}')
    
    plt.show()
```

<p>6. Написати функцію, яка буде виконувати пропорційне зменшення початкового зображення розміром N x M пікселей відповідно до вимоги, щоби зменешене зображення мало ширину K пікселей.</p>

```python
import cv2 as cv
import math
import matplotlib.pyplot as plt
import os

def resize_proportional(image, k_width):

    (height_N, width_M) = image.shape[:2]

    ratio = k_width / float(width_M)

    new_height = int(height_N * ratio)

    new_dimensions = (k_width, new_height)

    resized_image = cv.resize(image, new_dimensions, interpolation=cv.INTER_AREA)
    
    return resized_image

image_path = 'scr/123.png'
K = 300

image_rgb = cv.imread(image_path)

if image_rgb is None:
    print(f"Помилка: Не вдалося знайти зображення за шляхом: {image_path}")
else:
    image_resized_k = resize_proportional(image_rgb, K)
    
    print(f"Результати Завдання")
    print(f"Оригінальний розмір (M x N): {image_rgb.shape[1]} x {image_rgb.shape[0]}")
    print(f"Новий розмір (K x ...):   {image_resized_k.shape[1]} x {image_resized_k.shape[0]}")

    plt.imshow(cv.cvtColor(image_resized_k, cv.COLOR_BGR2RGB))
    plt.title(f'Зображення з новою шириною K = {K}px')
    plt.show()
```

<p>7. Зробити фото розвороту власної залікової книжки з перспективним викривленням на фоні письмового столу. Вирізати зону із заліковкою та виконати афінні та перспективні перетворення за трьома точками точками і візуалізовані. В кожному випадку результати мають бути записані на диск.</p>

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

zalikova_path = 'scr/1234.jpg' 
image = cv.imread(zalikova_path)

if image is None:
    print(f"Помилка: Не вдалося знайти фото за шляхом: {zalikova_path}")
else:
    print("Афінне перетворення")

    pts_input_affine = np.float32([
        [415, 340],
        [573, 342],
        [410, 540]
    ])

    pts_output_affine = np.float32([
        [150, 220],
        [450, 210],
        [140, 600]
    ])

    M_affine = cv.getAffineTransform(pts_input_affine, pts_output_affine)

    rows, cols, _ = image.shape
    image_affine = cv.warpAffine(image, M_affine, (cols, rows))

    cv.imwrite('scr/zalikova_affine_result.png', image_affine)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title("Оригінал")
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_affine, cv.COLOR_BGR2RGB))
    plt.title("Результат: Афінне (3 точки)")
    plt.show()
    
    print("Файл 'zalikova_affine_result.png' збережено.")
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

zalikova_path = 'scr/1234.jpg' 
image = cv.imread(zalikova_path)

if image is not None:
    print("\nПерспективне перетворення")
    
    pts_input_perspective = np.float32([
        [410, 340],
        [570, 340],
        [410, 550],
        [570, 550]
    ])

    output_width = 300
    output_height = 400

    pts_output_perspective = np.float32([
        [0, 0],             
        [output_width, 0],     
        [0, output_height],  
        [output_width, output_height]
    ])

    M_perspective = cv.getPerspectiveTransform(pts_input_perspective, pts_output_perspective)

    image_perspective = cv.warpPerspective(image, M_perspective, (output_width, output_height))

    cv.imwrite('scr/zalikova_perspective_result02.png', image_perspective)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title("Оригінал")
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_perspective, cv.COLOR_BGR2RGB))
    plt.title("Результат: Перспективне")
    plt.show()
    
    print("Файл 'zalikova_perspective_result02.png' збережено.")
```

<h2 style="font-weight:bold; font-size:22px;">Контрольні запитання</h2>

<ol>
    <li>У чому полягає сенс афінних перетворень?</li><br>
    <p>Сенс афінних перетворень полягає у тому, що паралельні лінії на вихідному зображенні залишаються паралельними і на трансформованому зображенні.
Вони є комбінацією лінійних перетворень (як-от масштабування, поворот, зсув) та трансляції (переміщення). Хоча лінії залишаються прямими, афінні перетворення, як правило, не зберігають кути між лініями або відстані між точками.</p>
    <li>Чи є оперції трансляції, масштабування і повороту афінними? Відповідь обгрунтувати.</li><br>
    <p>Так, є <br> Кожна з цих операцій відповідає головному правилу афінних перетворень — збереженню паралельності:</p>
    <p>Трансляція: Просто переміщує всі точки зображення на певний вектор (t_x, t_y). Паралельні лінії залишаються паралельними.</p>
    <p>Масштабування: Змінює розмір зображення, але паралельні лінії залишаються паралельними (лише змінюється відстань між ними).</p>
    <p>Поворот: Обертає зображення, але паралельні лінії також обертаються разом і залишаються паралельними.</p>
    <li>Чи є перспективні перетворення частинним випадком афінних перетворень? Відповідь обгрунтувати.</li><br>
    <p>Ні, не є. Насправді, все навпаки: афінне перетворення є частинним випадком перспективного.</p>
    <p>Афінні перетворення завжди зберігають паралельність ліній. Прямокутник після такого перетворення може стати паралелограмом, але його протилежні сторони залишаться паралельними.</p>
    <p>Перспективні перетворення не зберігають паралельність. Вони імітують те, як об'єкти виглядають у 3D-просторі (наприклад, рейки, що сходяться в точці на горизонті). Прямокутник, що зазнав перспективного викривлення (як на фото залікової книжки), стає трапецією, де паралельні в реальності сторони на зображенні вже не є паралельними.</p>
    <p>Також, для визначення афінного перетворення достатньо 3 пар точок, а для перспективного — 4.</p>
</ol>
