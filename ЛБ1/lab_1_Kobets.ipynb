<ol>
    <li>Що таке pip і conda та чим вони відрізняються між собою?</li><br>
    <p><strong>pip</strong> - це стандартний менеджер пакетів у Python, за допомогою якого можна встановлювати різні бібліотеки з PyPI.</p>
    <p><strong>conda</strong> - це теж менеджер пакетів, але він більш універсальний, бо вміє працювати не тільки з Python-бібліотеками, а й з іншими залежностями (наприклад, C-бібліотеками).
Основна різниця в тому, що pip ставить тільки Python-пакети, а conda створює цілі середовища з усіма потрібними компонентами.</p>
    <li>Як буде виглядати виконання пунктів 1-3 індивідуального завдання за допомогою Virtualenv?</li><br>

    <p>Потім активуємо його: myenv/Scripts/activate</p>
    <p>Встановити пакети: pip install numpy matplotlib opencv-python jupyter</p>
